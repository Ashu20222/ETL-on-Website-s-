# Extracting and Transforming mens batting data 

```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.cricbuzz.com/cricket-stats/icc-rankings/men/batting'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# Find the desired elements on the page
position1 = soup.find_all('div', class_='cb-col cb-col-16 cb-rank-tbl cb-font-16')
names1 = soup.find_all('a', class_='text-hvr-underline text-bold cb-font-16')
countries1 = soup.find_all('div', class_='cb-font-12 text-gray')
Rating1 = soup.find_all('div', class_='cb-col cb-col-17 cb-rank-tbl pull-right')

positions = []
# Extract the text from the elements
for ptn in position1:
    playerposition = int(ptn.text)
    positions.append(playerposition)

player_names = []
# Extract the text from the elements
for name in names1:
    player_name = name.text
    player_names.append(player_name)
    
country_names = []
# Extract the text from the elements
for country in countries1:
    countree = country.text
    country_names.append(countree)
    
Ratings = []
# Extract the text from the elements
for i in Rating1:
    rati = int(i.text)
    Ratings.append(rati)
    
```

```python
ICC_Cricket_Test_Rankings_Mens_Batting = pd.DataFrame({'positions': positions, 'player_names':player_names, 'country_names': country_names, 'Ratings':Ratings})
```

```python
test_batsmen = ICC_Cricket_Test_Rankings_Mens_Batting.iloc[0:94]
print(test_batsmen.head(10))
print(test_batsmen.tail(10))
```

```python
odi_batsmen = ICC_Cricket_Test_Rankings_Mens_Batting.iloc[94:193]
print(odi_batsmen.head(10))
print(odi_batsmen.tail(10))
```

```python
t20_batsmen = ICC_Cricket_Test_Rankings_Mens_Batting.iloc[193:]
print(t20_batsmen.head(10))
print(t20_batsmen.tail(10))
```

# Extracting and Transforming mens bowling data

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.cricbuzz.com/cricket-stats/icc-rankings/men/bowling'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# Find the div elements that contain the player positions, names, countries, and ratings
positions2 = soup.find_all('div', class_='cb-col cb-col-16 cb-rank-tbl cb-font-16')
names2 = soup.find_all('a', class_='text-hvr-underline text-bold cb-font-16')
countries2 = soup.find_all('div', class_='cb-font-12 text-gray')
ratings2 = soup.find_all('div', class_='cb-col cb-col-17 cb-rank-tbl pull-right')

# Extract the data from the elements and store it in separate lists
Bowler_positions = []
for ptn in positions2:
    player_position = int(ptn.text)
    Bowler_positions.append(player_position)

Bowler_names = []
for name in names2:
    player_name = name.text
    Bowler_names.append(player_name)

Bowler_country_names = []
for country in countries2:
    country_name = country.text
    Bowler_country_names.append(country_name)

Bowler_ratings = []
for rating in ratings2:
    player_rating = int(rating.text)
    Bowler_ratings.append(player_rating)
```

```python
ICC_Cricket_Test_Rankings_Mens_Bowling = pd.DataFrame({'Bowler_positions': Bowler_positions, 'Bowler_names':Bowler_names, 'Bowler_country_names': Bowler_country_names, 'Bowler_ratings':Bowler_ratings})
```

# Extracting and Transforming mens all rounders data

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.cricbuzz.com/cricket-stats/icc-rankings/men/all-rounder'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# Find the div elements that contain the player positions, names, countries, and ratings
positions3 = soup.find_all('div', class_='cb-col cb-col-16 cb-rank-tbl cb-font-16')
names3 = soup.find_all('a', class_='text-hvr-underline text-bold cb-font-16')
countries3 = soup.find_all('div', class_='cb-font-12 text-gray')
ratings3 = soup.find_all('div', class_='cb-col cb-col-17 cb-rank-tbl pull-right')

# Extract the data from the elements and store it in separate lists
All_Rounders_positions = []
for ptn in positions3:
    player_position = int(ptn.text)
    All_Rounders_positions.append(player_position)

All_Rounders_names = []
for name in names3:
    player_name = name.text
    All_Rounders_names.append(player_name)

All_Rounders_country_names = []
for country in countries3:
    country_name = country.text
    All_Rounders_country_names.append(country_name)

All_Rounders_ratings = []
for rating in ratings3:
    player_rating = int(rating.text)
    All_Rounders_ratings.append(player_rating)
```

```python
ICC_Cricket_Test_Rankings_Mens_All_Rounders = pd.DataFrame({'All_Rounders_positions': All_Rounders_positions, 'All_Rounders_names':All_Rounders_names, 'All_Rounders_country_names': All_Rounders_country_names, 'All_Rounders_ratings':All_Rounders_ratings},index=None)
```
